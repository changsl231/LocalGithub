# 24. NumPy Matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(-100,100)
#y = 2 * x + 5
#y = 1/(1+np.e**-x)
y = 1/(1+np.power(np.e,-x))
plt.title("S")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
plt.title("S")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,'.')
plt.show()
```

绘制正弦波

```python
#计算正弦曲线上点的x和y坐标
x = np.arange(0,3 * np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")
#使用mayplotlib来绘点
plt.plot(x,y)
plt.show()
```

subplot()
subplot() 函数允许你在同一图中绘制不同的东西。

```python
#计算正弦和余弦曲线上的点的x和y坐标
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
#建立subplot网格，高为2，宽为1
#激活第一个subplot
plt.subplot(2,1,1)
#绘制第一个图象
plt.plot(x,y_sin)
plt.title("Sine")
#绘制第二个图像
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")
#展示图像
plt.show()
```

bar()
pyplot 子模块提供 bar() 函数来生成条形图。 

```python
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.bar(x,y,align = 'center')
plt.bar(x2,y2,color = 'g',align = 'center')
plt.title("Bar graph")
plt.ylabel("Y axis")
plt.ylabel("X axis")
plt.show()
```

```python
#激活第一个subplot
plt.subplot(2,1,1)
plt.bar(x,y,align = 'center')
plt.title("Bar graph")
plt.ylabel("Y1 axis")
plt.ylabel("X1 axis")
#激活第一个subplot
plt.subplot(2,1,2)
plt.bar(x2,y2,color = 'g',align = 'center')
plt.title("Bar graph")
plt.ylabel("Y2 axis")
plt.ylabel("X2 axis")
plt.show()
```

numpy.histogram()
numpy.histogram() 函数是数据的频率分布的图形表示。 水平尺寸相等的矩形对应于类间隔，称为 bin，变量 height 对应于频率。
numpy.histogram()函数将输入数组和 bin 作为两个参数。 bin 数组中的连续元素用作每个 bin 的边界。

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,96])
hist,bins=np.histogram(a,bins = [0,20,40,60,80,100])
print(hist)
print(bins)
```

# Matplotlib 教程

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi,np.pi,256,endpoint = True)
C,S = np.cos(X),np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

```python
#导入matplotlib的所有内容（numpy 可以用np这个名字来使用）
from pylab import *

#创建一个8*6点（point)的图，并设置分辨率为80
figure(figsize=(8,6), dpi=80)
#创建一个新的1*1的子图
subplot(1,1,1)

#绘制余弦曲线
plot(X,C,color="bLue",linewidth=1.0,linestyle="-")
#绘制正弦曲线
plot(X,S,color="green",linewidth=1.0,linestyle="-")
#设置x轴标签
xlabel("x")
#设置横轴上下限
xlim(-4.0,4.0)
#设置y轴标签
ylabel("y")
#设置纵轴记号
yticks(np.linspace(-1,1,11,endpoint=True))
#在屏幕上显示
show()
```

设置图片边界

```python
xmin,xmax = X.min(),X.max()
ymin,ymax = C.min(),C.max()

dx = (xmax - xmin)*0.2
dy = (ymax - ymin)*0.2

xlim(xmin-dx,xmax+dx)
ylim(ymin-dx,ymax+dx)
plot(X,C)
plot(X,S)
show()
```

```python
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
yticks([-1,0,+1])
plot(X,C,color='blue',linewidth=2.5,linestyle='-',label="cosine")
plot(X,S,color='red' ,linewidth=2.5,linestyle='-',label="sine")
legend(loc='upper left')
show()
```

```python
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
yticks([-1,0,+1])

t = 2*np.pi/3
plot(X,C,color='blue',linewidth=2.5,linestyle='-',label="cosine")
plot([t,t],[0,np.cos(t)],color="blue",linewidth=2.5,linestyle='--')
scatter([t],[np.cos(t)],50,color = 'yellow')

annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
         xy=(t, np.cos(t)), xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plot(X,S,color='red' ,linewidth=2.5,linestyle='-',label="sine")
plot([t,t],[0,np.sin(t)],color="red",linewidth=2.5,linestyle='--')
scatter([t],[np.sin(t)],50,color = 'green')

annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
         xy=(t, np.sin(t)), xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))


for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
legend(loc='upper left')
show()
```

```python
from pylab import *
import matplotlib.gridspec as gridspec

X = np.linspace(-np.pi,np.pi,256,endpoint = True)
C,S = np.cos(X),np.sin(X)
G = gridspec.GridSpec(3, 3)

axes_1 = subplot(G[0, :])
xticks([]), yticks([])
#text(0.5,0.5, 'Axes 1',ha='center',va='center',size=24,alpha=.5)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
yticks([-1,0,+1])

t = 2*np.pi/3
plot(X,C,color='blue',linewidth=2.5,linestyle='-',label="cosine")
plot([t,t],[0,np.cos(t)],color="blue",linewidth=2.5,linestyle='--')
scatter([t],[np.cos(t)],50,color = 'yellow')

annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
         xy=(t, np.cos(t)), xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plot(X,S,color='red' ,linewidth=2.5,linestyle='-',label="sine")
plot([t,t],[0,np.sin(t)],color="red",linewidth=2.5,linestyle='--')
scatter([t],[np.sin(t)],50,color = 'green')

annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
         xy=(t, np.sin(t)), xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))


for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
legend(loc='upper left')
show()

axes_2 = subplot(G[1,:-1])
xticks([]), yticks([])
text(0.5,0.5, 'Axes 2',ha='center',va='center',size=24,alpha=.5)

axes_3 = subplot(G[1:, -1])
xticks([]), yticks([])
text(0.5,0.5, 'Axes 3',ha='center',va='center',size=24,alpha=.5)

axes_4 = subplot(G[-1,0])
xticks([]), yticks([])
text(0.5,0.5, 'Axes 4',ha='center',va='center',size=24,alpha=.5)

axes_5 = subplot(G[-1,-2])
xticks([]), yticks([])
text(0.5,0.5, 'Axes 5',ha='center',va='center',size=24,alpha=.5)

#plt.savefig('../figures/gridspec.png', dpi=64)
show()
```

```python
from pylab import *

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

ylim(-1.25,+1.25)
show()
```

```python
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black')
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

show()
```
